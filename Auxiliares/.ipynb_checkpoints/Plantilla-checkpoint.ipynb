```python
# Otra estética con difuminado
# Superponer gráficas para no ver lo límites de las curvas tan acentuados
sns.set_style('darkgrid')
k1 = sns.kdeplot(data.Usuarios_de_Internet , data.Tasa_de_Natalidad , \
                shade=True , shade_lowest=False , cmap='Reds')

k2 = sns.kdeplot(data.Usuarios_de_Internet , data.Tasa_de_Natalidad , \
                 cmap='Reds') # Límites de cada capa por encima de la simbra
```

```python
# Inferno
k2 = sns.kdeplot(data.Usuarios_de_Internet , data.Tasa_de_Natalidad ,\
                shade=True , shadelowest=True , cmap='inferno' )
                # Estilo de la sombra
k2b = sns.kdeplot(data.Usuarios_de_Internet , data.Tasa_de_Natalidad ,\
                 cmap='cool') # Estilo del delineado
```

```python
# Multiples estéticas en un subplot de distintas librerías
sns.set_style('darkgrid')
f , axes = plt.subplots(2,2 , figsize=(15,15))

# Gráfico [0,0]
k1 = sns.kdeplot(peliculas.PresupuestoMillones, peliculas.RatingsAudencia,\
                 shade=True , shadelowest=True , cmap='Blues' , \
                 ax=axes[0,0])
k1b = sns.kdeplot(peliculas.PresupuestoMillones, peliculas.RatingsAudencia,\
                 cmap='Blues' , ax=axes[0,0])

# Gráfico [0,1]
k2 = sns.kdeplot(peliculas.PresupuestoMillones, peliculas.RatingsCriticos,\
                 shade=True , shadelowest=True , cmap='inferno' ,\
                 ax=axes[0,1]) # Estilo de la sombra
k2b = sns.kdeplot(peliculas.PresupuestoMillones, peliculas.RatingsCriticos,\
                 cmap='cool' , ax=axes[0,1]) # Estilo del delineado

# Gráfico [1,0]
z = sns.violinplot(data=peliculas[peliculas.Genero=='Drama'] , x='Año' , \
                   y='RatingsCriticos' , ax=axes[1,0])

# Gráfico [1,1]
k4 = sns.kdeplot(peliculas.RatingsCriticos , peliculas.RatingsAudencia, \
                shade=True , shade_lowest=False , cmap='Reds' , \
                 ax=axes[1,1])
k4b = sns.kdeplot(peliculas.RatingsCriticos , peliculas.RatingsAudencia, \
                cmap='Reds', ax=axes[1,1])

k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
# Generar etiquetas a partir del datafrane para un histograma

sns.set_style('darkgrid')
lista1 = list()
mis_etiquetas = list()

for gen in peliculas.Genero.cat.categories:
    lista1.append(peliculas[peliculas.Genero == \
    gen].PresupuestoMillones)
    mis_etiquetas.append(gen)

h = plt.hist(lista1 , bins = 30 , stacked = True , \
             rwidth = 1 , label = mis_etiquetas)
plt.legend()
plt.show()
```

```python
# Estilos complementarios dentro del histograma
sns.set_style('whitegrid')
lista1 = list()
mis_etiquetas = list()

for gen in peliculas.Genero.cat.categories:
    lista1.append(peliculas[peliculas.Genero == \
    gen].PresupuestoMillones)
    mis_etiquetas.append(gen)
    
fig , ax = plt.subplots(1,1)
fig.set_size_inches(11.7 , 8.27) # Tamaño de una hoja A4

h = plt.hist(lista1 , bins = 30 , stacked = True , rwidth = 1 ,\
             edgecolor='none' , label = mis_etiquetas)
plt.title('Distribución de Presupuestos para las Películas',\
         fontsize=30 , color='Darkblue' , fontname="Console")
plt.ylabel("Número de Películas" , fontsize=20 , color="Red")
plt.xlabel("Presupuesto" , fontsize=20 , color="Green")
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.legend(frameon=True , fancybox=True , framealpha=1 ,\
           shadow=True , prop={'size':20})
plt.show()
```

```python
data = np.array([[25791017 , 34923129 , 48225238 , 66846833 , 81249645 , 97483412 , 112336538], # Número asociado a la población total
                [1950 , 1960 , 1970 , 1980 , 1990 , 2000 , 2010]]) # Año de la medida de la población total. 
```
